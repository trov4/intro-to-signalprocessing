## What does a sin wave look and sound like? 
Let's look at the properties of a sin wave and how to interact with them.

```python
import numpy as np
import matplotlib.pylab as plt
from scipy import fft,ifft
import wavio

def plot_fft(x, sig_f, title, N):
    plt.plot(x, 2.0*fs * np.abs(sig_f[:N//2]))
    plt.title(title)
    plt.show()

# generate a sin wave using numpy
fs = 44100
freq = 5000
N = fs * 10
samples = np.linspace(0, 10, N) # last value is time * frequency
sig_1 = np.sin(2*np.pi*samples* freq)

frac_sec = int(fs / 1000)
# show time domain
plt.plot(samples[:frac_sec], sig_1[:frac_sec])
plt.title("Time domain of 1000 Hz signal")
plt.show()

# show frequency domain
sig_1f = fft(sig_1)
x = np.linspace(0, 1.0/(2.0*1/fs), N//2)
plot_fft(x, sig_1f, "Frequency domain of 1000 Hz signal", N)

# output audio wave
wavio.write("5k_sinwave.wav", sig_1, fs, sampwidth=2)
print("Saved file 5k_sinwave.wav")
```

## What happens when theres more than 1 sin wave?

```python
# generate 1KHz signal and add to 5KHz signal
freq_2 = 1000
sig_2 = np.sin(2*np.pi*samples*freq_2)
sig_1_2 = sig_1 + sig_2 # combine waves here

# plot combimned signal
plt.plot(samples[:frac_sec], sig_1_2[:frac_sec])
plt.title("Time domain of combined signals")
plt.show()

# show frequency domain
sig_1_2_f = fft(sig_1_2)
plot_fft(x, sig_1_2_f, "Frequency Domain of Combined Signals", N)

# save audio file
wavio.write("combined_sinwaves.wav", sig_1_2, fs, sampwidth=2)
print("saved combined sinwaves")
```

## Now let's see how a song looks

```python
# readin audio file
song_name = "Fergessen_BackFromTheStart.wav"

# reads in stereo, convert to mono
song_wav = wavio.read(song_name) # stereo
print(song_wav)
song = song_wav.data[:,0]# + song_wav.data[:, 1] # mono

# show time domain
plt.plot(samples[:frac_sec], song[:frac_sec])
plt.title("Time domain of a song")
plt.show()

song_f = fft(song)
N_s = fs * 18
x_s = np.linspace(0, 1.0/(2.0*1/fs), N_s//2)
plot_fft(x_s, song_f, "Frequency Domain of a song", N_s)
```

## Is it possible to add a filter?

```python
from math import log
fc = .6 # fraction of fs to be cutoff
b = .6 # transition abnd as a franction of fs
N_filt = int(np.ceil((4/b)))
if not N_filt%2: N_filt+=1
n = np.arange(N_filt)

# create a lowpass filter
h = np.sinc(2 * fc * (n - (N_filt) / 2))
w = np.blackman(N_filt)
h = h * w
h = h/np.sum(h)

# create a lowpass filter from the high pass filter
h = -h
h[(N_filt - 1)//2] += 1
plt.plot(h)
plt.show()

# apply filter through convolution
song_filt = np.convolve(song, h)
plot_fft(x_s, song_f, "Song before filter", N_s)
plot_fft(x_s, fft(song_filt), "Song with filter applied", N_s)

# save wav file to listen to
from scipy import ifft #inverse fast fourier transform
wavio.write("filtered_" + song_name, song_filt, fs, sampwidth=4)
print("saved filtered song")
```
